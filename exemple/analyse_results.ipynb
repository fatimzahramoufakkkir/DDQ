```python
cas=0
import os
dpbound=[]
E0=np.genfromtxt("real_sin2/t2/liant_nompi_3h/E0value.dat")
for cas in range(len(E0[:,3])):
    diss=np.genfromtxt("real_sin2/t2/diss_nompi_3h/pbound."+str(10000+cas))
    liand=np.genfromtxt("real_sin2/t2/liant_nompi_3h/pbound."+str(10000+cas))
    print("cas "+str(cas)+", liand[-1,1]="+str(liand[-1,1])+", diss[-1,1]="+str(diss[-1,1]))
    plt.plot(liand[:,0],liand[:,3])
    plt.plot(liand[:,0],liand[:,2],label="Bound case")
    plt.plot(diss[:,0],diss[:,2],label="Dissociative case")
    plt.legend(loc=0)
    plt.show()
    plt.close()
    plt.plot(liand[:,0],liand[:,1],label="liand")
    plt.plot(diss[:,0],diss[:,1],label="diss")
    plt.legend()
    plt.show()
    plt.close()
    dpbound.append(liand[-1,1]-diss[-1,1])
    cas=cas+1
E0wcm2=E0[:,3]**2*3.5094475E16
plt.plot(E0wcm2,dpbound)
plt.xlabel("test")
plt.show()
```
