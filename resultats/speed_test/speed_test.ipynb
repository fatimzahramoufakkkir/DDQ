```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# real time
# 48 9m45.406s
# 24 proc 14m30.734s
# 12 proc  28m29.723s
# 6 proc 55m51.364s
# 3 procs 111m20.507s
# 1 proc 333m29.889s
#
proctime1=(333*60+29)


core=[1,3,6,12,24,48]
time=np.array([333*60+30,111*60+21,55*60+51,28*60+30,14*60+31,9*60+45])/1000
ax1=plt.subplot(1, 1, 1)
plt.plot(core,time)
plt.xlabel("Number of cores")
plt.ylabel("Running time (k sec.)")
#plt.show()
#plt.close()
ax2 = ax1.twinx()
speedup3=proctime1/(111*60+21)
speedup6=proctime1/(55*60+51)
speedup12=proctime1/(28*60+30)
speedup24=proctime1/(14*60+31)
speedup48=proctime1/(9*60+45)
print("speedup3="+str(speedup3))
print("speedup6="+str(speedup6))
print("speedup12="+str(speedup12))
print("speedup24="+str(speedup24))
print("speedup48="+str(speedup48))

ax2.plot(core,[100,speedup3/3*100,speedup6/6*100,speedup12/12*100,speedup24/24*100,speedup48/48*100],'r')
ax2.set_ylabel(r'Efficienty (%)')#,fontsize=16)
plt.savefig("DDQ_efficienty.png")
plt.show()
plt.close()

```
